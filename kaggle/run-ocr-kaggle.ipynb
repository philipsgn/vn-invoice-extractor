```python
%%writefile run_ocr_kaggle.py
# -*- coding: utf-8 -*-
"""
run_ocr_kaggle.py  —  Production | Colab/Kaggle T4/P100 GPU
======================================================
PaddleOCR (Detection, rec=True) + VietOCR (Recognition, GPU batch)
Auto-zip output folder sau khi OCR xong.

FIXES:
  [FIX-1] tokens: List[{"text","bbox"}] — eliminates fallback warnings downstream
  [FIX-2] Reading order: snap y → row band, tránh đảo thứ tự token cùng dòng
  [FIX-3] numpy ambiguous truth value: dùng len() thay vì `not` trên array
  [FIX-4] Checkpoint atomic write: dùng os.replace() tránh corrupt khi kill
  [FIX-5] RAM leak: del loaded + del img_bgr sau mỗi chunk
  [FIX-6] ErrorLogger flush định kỳ: tránh OOM + mất log khi kill
  [FIX-7] gpu_clear_every: move vào inner loop, trigger đúng mỗi N ảnh
======================================================
"""

import argparse
import gc
import json
import os
import time
import zipfile
from concurrent.futures import ThreadPoolExecutor, as_completed
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Dict, List, Optional, Sequence, Tuple

import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm

if not hasattr(Image, "ANTIALIAS"):
    Image.ANTIALIAS = Image.Resampling.LANCZOS


# ======================================================
# CONSTANTS
# ======================================================
DEFAULT_BATCH_SIZE      = 32
DEFAULT_WORKERS         = 2
DEFAULT_GPU_CLEAR_EVERY = 500
LINE_HEIGHT_ESTIMATE    = 30


# ======================================================
# UTILITIES
# ======================================================

def _ensure_dir(path: Path) -> None:
    path.mkdir(parents=True, exist_ok=True)


def _clip(val: int, lo: int, hi: int) -> int:
    return max(lo, min(hi, val))


def _quad_to_xyxy(quad: Sequence[Sequence[float]]) -> Tuple[int, int, int, int]:
    xs = [p[0] for p in quad]
    ys = [p[1] for p in quad]
    return (
        int(np.floor(min(xs))), int(np.floor(min(ys))),
        int(np.ceil(max(xs))),  int(np.ceil(max(ys))),
    )


def _sanitize_text(s: str) -> str:
    return " ".join((s or "").strip().split())


def _reading_order_key(
    line: Dict[str, Any],
    line_height: int = LINE_HEIGHT_ESTIMATE,
) -> Tuple[int, int]:
    x1, y1 = line["bbox"][0], line["bbox"][1]
    return (y1 // line_height, x1)


def _split_text_to_token_dicts(
    text: str,
    x1: int, y1: int, x2: int, y2: int,
) -> List[Dict[str, Any]]:
    words = text.split()
    if not words:
        return []
    line_width = x2 - x1
    if line_width <= 0:
        return [{"text": w, "bbox": [x1, y1, x2, y2]} for w in words]
    char_counts = [max(1, len(w)) for w in words]
    total_chars = sum(char_counts)
    result      = []
    cursor_x    = x1
    for i, (word, n_chars) in enumerate(zip(words, char_counts)):
        wx2 = (
            x2 if i == len(words) - 1
            else cursor_x + int(line_width * n_chars / total_chars)
        )
        result.append({"text": word, "bbox": [cursor_x, int(y1), wx2, int(y2)]})
        cursor_x = wx2
    return result


# ======================================================
# AUTO ZIP
# ======================================================

def zip_output(output_dir: Path) -> Optional[Path]:
    zip_path = output_dir.parent / f"{output_dir.name}.zip"
    files    = [f for f in output_dir.rglob("*") if f.is_file()]

    if not files:
        print("⚠️  No files to zip.")
        return None

    print(f"\n📦 Zipping {len(files)} files → {zip_path.name} ...")
    t0 = time.time()

    with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
        for f in tqdm(files, desc="Zipping", unit="file", dynamic_ncols=True):
            zf.write(f, f.relative_to(output_dir))

    elapsed = time.time() - t0
    size_mb = zip_path.stat().st_size / 1e6
    print(f"✅ Zip done  →  {zip_path}")
    print(f"   Size : {size_mb:.1f} MB  |  Time : {elapsed:.1f}s")
    print(f"   👉 Kaggle → tab Output (phải) → download {zip_path.name}")
    return zip_path


# ======================================================
# DATA STRUCTURES
# ======================================================

@dataclass
class DetectedLine:
    quad: List[List[float]]
    xyxy: Tuple[int, int, int, int]


# ======================================================
# CHECKPOINT & ERROR LOGGER
# ======================================================

class CheckpointManager:
    def __init__(self, path: Path):
        self.path = path
        self.processed: set = self._load()

    def _load(self) -> set:
        if self.path.exists():
            try:
                return set(
                    json.loads(self.path.read_text(encoding="utf-8"))
                    .get("processed", [])
                )
            except Exception:
                return set()
        return set()

    def save(self, filename: str) -> None:
        self.processed.add(filename)
        try:
            # [FIX-4] Atomic write: ghi ra .tmp trước, rename sau
            # → tránh corrupt checkpoint nếu process bị kill giữa chừng
            tmp_path = self.path.with_suffix(".tmp")
            tmp_path.write_text(
                json.dumps(
                    {"processed": list(self.processed), "count": len(self.processed)},
                    ensure_ascii=False, indent=2,
                ),
                encoding="utf-8",
            )
            os.replace(tmp_path, self.path)  # atomic trên Linux
        except Exception as e:
            print(f"⚠️ Checkpoint save failed: {e}")

    def is_done(self, filename: str) -> bool:
        return filename in self.processed


class ErrorLogger:
    def __init__(self, path: Path):
        self.path   = path
        self.errors: List[Dict] = []

    def log(self, filename: str, error: str) -> None:
        self.errors.append({
            "filename":  filename,
            "error":     str(error),
            "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
        })
        # [FIX-6] Flush ra disk mỗi 100 errors
        # → tránh OOM khi có nhiều lỗi + không mất log nếu bị kill
        if len(self.errors) % 100 == 0:
            self._flush()

    def _flush(self) -> None:
        try:
            existing = []
            if self.path.exists():
                try:
                    existing = json.loads(self.path.read_text(encoding="utf-8"))
                except Exception:
                    existing = []
            existing.extend(self.errors)
            tmp_path = self.path.with_suffix(".errtmp")
            tmp_path.write_text(
                json.dumps(existing, ensure_ascii=False, indent=2),
                encoding="utf-8",
            )
            os.replace(tmp_path, self.path)
            self.errors.clear()
        except Exception:
            pass  # giữ errors trong memory nếu flush fail

    def save(self) -> None:
        if self.errors:
            self._flush()
        if self.path.exists():
            try:
                count = len(json.loads(self.path.read_text(encoding="utf-8")))
                print(f"📝 {count} errors → {self.path}")
            except Exception:
                pass


# ======================================================
# OCR RUNNER
# ======================================================

class OCRRunner:
    def __init__(
        self,
        device:             str  = "cuda",
        batch_size:         int  = DEFAULT_BATCH_SIZE,
        paddle_lang:        str  = "vi",
        vietocr_model:      str  = "vgg_transformer",
        vietocr_beamsearch: bool = False,
        verbose:            bool = False,
    ) -> None:
        self.device     = device
        self.batch_size = batch_size
        self.verbose    = verbose
        self._use_gpu   = (device == "cuda")

        print(f"🔧 Device: {device.upper()}  |  Batch size: {batch_size}")
        print("🔄 Loading PaddleOCR...")
        self._paddle  = self._init_paddle(paddle_lang)
        print("🔄 Loading VietOCR...")
        self._vietocr = self._init_vietocr(vietocr_model, vietocr_beamsearch)
        print("✅ Models loaded\n")

    def _init_paddle(self, lang: str) -> Any:
        try:
            from paddleocr import PaddleOCR
        except ImportError:
            raise RuntimeError("pip install paddlepaddle-gpu paddleocr")
        return PaddleOCR(
            lang=lang,
            use_gpu=self._use_gpu,
            det=True,
            rec=True,            # rec=True → Python list output, stable mọi version
            use_angle_cls=False, # Hóa đơn VN không có text xoay
            show_log=self.verbose,
        )

    def _init_vietocr(self, model_name: str, beamsearch: bool) -> Any:
        try:
            from vietocr.tool.config import Cfg
            from vietocr.tool.predictor import Predictor
        except ImportError:
            raise RuntimeError("pip install vietocr")
        cfg = Cfg.load_config_from_name(model_name)
        cfg["device"]                  = self.device
        cfg["predictor"]["beamsearch"] = beamsearch
        return Predictor(cfg)

    def load_image(self, path: Path) -> Tuple[np.ndarray, int, int]:
        img = cv2.imread(str(path))
        if img is None:
            raise ValueError(f"Cannot read: {path}")
        h, w = img.shape[:2]
        return img, w, h

    def detect(self, image_path: Path, img: np.ndarray) -> List[DetectedLine]:
        h, w = img.shape[:2]
        raw  = self._paddle.ocr(str(image_path))

        # [FIX-3] Dùng len() thay vì `not` — tránh numpy ambiguous truth value
        if raw is None or len(raw) == 0 or raw[0] is None or len(raw[0]) == 0:
            return []

        lines: List[DetectedLine] = []
        for item in raw[0]:
            # rec=True: item = [quad, (text, conf)]
            if item is None or len(item) < 2:
                continue
            quad = item[0]
            # [FIX-3] Dùng len() thay vì `not quad`
            if quad is None or len(quad) == 0:
                continue
            x1, y1, x2, y2 = _quad_to_xyxy(quad)
            x1 = _clip(x1, 0, w - 1)
            y1 = _clip(y1, 0, h - 1)
            x2 = _clip(x2, 0, w)
            y2 = _clip(y2, 0, h)
            if x2 <= x1 or y2 <= y1:
                continue
            lines.append(DetectedLine(
                quad=[[float(p[0]), float(p[1])] for p in quad],
                xyxy=(x1, y1, x2, y2),
            ))
        return lines

    def recognize_batch(self, crops_bgr: List[np.ndarray]) -> List[str]:
        if not crops_bgr:
            return []

        pil_images, valid_mask = [], []
        for c in crops_bgr:
            if c is not None and c.size > 0:
                pil_images.append(
                    Image.fromarray(cv2.cvtColor(c, cv2.COLOR_BGR2RGB))
                )
                valid_mask.append(True)
            else:
                valid_mask.append(False)

        if not pil_images:
            return [""] * len(crops_bgr)

        try:
            texts = self._vietocr.predict_batch(pil_images)
        except Exception as e:
            if self.verbose:
                print(f"  ⚠️  Batch failed ({e}), fallback single...")
            texts = []
            for img in pil_images:
                try:
                    texts.append(self._vietocr.predict(img))
                except Exception:
                    texts.append("")

        result, valid_idx = [], 0
        for is_valid in valid_mask:
            if is_valid:
                raw = texts[valid_idx] if valid_idx < len(texts) else ""
                result.append(_sanitize_text(raw))
                valid_idx += 1
            else:
                result.append("")
        return result

    def process_image(
        self,
        image_path: Path,
        img_bgr:    np.ndarray,
        width:      int,
        height:     int,
    ) -> Dict[str, Any]:
        detections = self.detect(image_path, img_bgr)
        if not detections:
            return {
                "image": image_path.name, "width": width,
                "height": height, "num_lines": 0, "lines": [],
            }

        crops, valid_dets = [], []
        for det in detections:
            x1, y1, x2, y2 = det.xyxy
            crop = img_bgr[y1:y2, x1:x2]
            if crop.size > 0:
                crops.append(crop)
                valid_dets.append(det)

        all_texts: List[str] = []
        for i in range(0, len(crops), self.batch_size):
            all_texts.extend(self.recognize_batch(crops[i: i + self.batch_size]))

        lines_raw = []
        for det, text in zip(valid_dets, all_texts):
            if not text:
                continue
            x1, y1, x2, y2 = det.xyxy
            lines_raw.append({
                "text":       text,
                "tokens":     _split_text_to_token_dicts(text, x1, y1, x2, y2),
                "bbox":       [int(x1), int(y1), int(x2), int(y2)],
                "confidence": 1.0,
            })

        if lines_raw:
            line_h = max(10, int(np.median(
                [r["bbox"][3] - r["bbox"][1] for r in lines_raw]
            )))
        else:
            line_h = LINE_HEIGHT_ESTIMATE
        lines_raw.sort(key=lambda r: _reading_order_key(r, line_h))

        for i, r in enumerate(lines_raw, start=1):
            r["id"] = i

        return {
            "image":     image_path.name,
            "width":     int(width),
            "height":    int(height),
            "num_lines": len(lines_raw),
            "lines":     lines_raw,
        }


# ======================================================
# MAIN LOOP
# ======================================================

def run_folder(
    input_dir:          Path,
    output_dir:         Path,
    device:             str   = "cuda",
    batch_size:         int   = DEFAULT_BATCH_SIZE,
    workers:            int   = DEFAULT_WORKERS,
    paddle_lang:        str   = "vi",
    vietocr_model:      str   = "vgg_transformer",
    vietocr_beamsearch: bool  = False,
    resume:             bool  = True,
    gpu_clear_every:    int   = DEFAULT_GPU_CLEAR_EVERY,
    auto_zip:           bool  = True,
    verbose:            bool  = False,
) -> None:
    _ensure_dir(output_dir)
    checkpoint   = CheckpointManager(output_dir / "_checkpoint.json") if resume else None
    error_logger = ErrorLogger(output_dir / "_errors.json")

    runner = OCRRunner(
        device=device,
        batch_size=batch_size,
        paddle_lang=paddle_lang,
        vietocr_model=vietocr_model,
        vietocr_beamsearch=vietocr_beamsearch,
        verbose=verbose,
    )

    image_paths = sorted(input_dir.glob("*.png")) + sorted(input_dir.glob("*.jpg"))
    if not image_paths:
        print(f"❌ No images found in {input_dir}")
        return

    print(f"📂 Found {len(image_paths)} images")

    if checkpoint and checkpoint.processed:
        before = len(image_paths)
        image_paths = [p for p in image_paths if not checkpoint.is_done(p.name)]
        skipped = before - len(image_paths)
        if skipped:
            print(f"⏭️  Resume: {skipped} done, {len(image_paths)} remaining")

    if not image_paths:
        print("✅ All images already processed!")
        if auto_zip:
            zip_output(output_dir)
        return

    stats = {"total": len(image_paths), "success": 0, "failed": 0, "lines": 0}
    start = time.time()

    def _load(path: Path):
        try:
            img, w, h = runner.load_image(path)
            return path, img, w, h, None
        except Exception as e:
            return path, None, 0, 0, str(e)

    print(f"\n🚀 OCR  device={device}  batch={batch_size}  workers={workers}\n")

    CHUNK           = max(batch_size * 4, 64)
    processed_count = 0

    with tqdm(total=len(image_paths), desc="OCR", unit="img", dynamic_ncols=True) as pbar:
        for chunk_start in range(0, len(image_paths), CHUNK):
            chunk = image_paths[chunk_start: chunk_start + CHUNK]

            # Pre-load ảnh song song (I/O overlap)
            loaded = []
            with ThreadPoolExecutor(max_workers=workers) as pool:
                futures = {pool.submit(_load, p): p for p in chunk}
                for fut in as_completed(futures):
                    loaded.append(fut.result())

            path_order = {p: i for i, p in enumerate(chunk)}
            loaded.sort(key=lambda x: path_order.get(x[0], 0))

            for img_path, img_bgr, w, h, err in loaded:
                if err:
                    stats["failed"] += 1
                    error_logger.log(img_path.name, err)
                    pbar.update(1)
                    continue

                try:
                    result = runner.process_image(img_path, img_bgr, w, h)
                    out_path = output_dir / f"{img_path.stem}.json"
                    out_path.write_text(
                        json.dumps(result, ensure_ascii=False, indent=2),
                        encoding="utf-8",
                    )
                    if checkpoint:
                        checkpoint.save(img_path.name)
                    stats["success"] += 1
                    stats["lines"]   += result["num_lines"]
                    processed_count  += 1
                except Exception as e:
                    stats["failed"] += 1
                    error_logger.log(img_path.name, str(e))
                    if verbose:
                        print(f"\n  ❌ {img_path.name}: {e}")
                finally:
                    # [FIX-5] Giải phóng RAM ngay sau khi process xong mỗi ảnh
                    if img_bgr is not None:
                        del img_bgr

                pbar.update(1)
                pbar.set_postfix(
                    ok=stats["success"],
                    fail=stats["failed"],
                    lines=stats["lines"],
                )

                # [FIX-7] GPU clear trong inner loop — trigger đúng mỗi N ảnh
                if device == "cuda" and processed_count > 0 and processed_count % gpu_clear_every == 0:
                    try:
                        import torch
                        torch.cuda.empty_cache()
                        gc.collect()
                    except Exception:
                        pass

            # [FIX-5] Giải phóng toàn bộ chunk RAM sau khi xử lý xong
            del loaded
            gc.collect()

    # ── Summary ───────────────────────────────────────────────
    elapsed = time.time() - start
    error_logger.save()

    speed = stats["success"] / elapsed if elapsed > 0 else 0
    print("\n" + "=" * 60)
    print("📊  OCR COMPLETE")
    print("=" * 60)
    print(f"  ✅ Success  : {stats['success']:,} / {stats['total']:,}")
    print(f"  ❌ Failed   : {stats['failed']:,}")
    print(f"  📝 Lines    : {stats['lines']:,}")
    print(f"  ⏱️  Time     : {elapsed:.1f}s  ({elapsed/60:.1f} min)")
    if speed > 0:
        print(f"  🚀 Speed    : {speed:.1f} img/s  (~{1/speed:.2f}s/img)")
    print(f"  📂 Output   : {output_dir}")
    if stats["failed"] > 0:
        print(f"  ⚠️  Errors  : {output_dir / '_errors.json'}")
    print("=" * 60)

    # ── Auto zip liền mạch sau OCR ────────────────────────────
    if auto_zip:
        zip_output(output_dir)


# ======================================================
# CLI
# ======================================================

def main() -> None:
    p = argparse.ArgumentParser(
        description="Vietnamese Invoice OCR — Colab/Kaggle GPU Production",
        formatter_class=argparse.RawDescriptionHelpFormatter,
    )
    p.add_argument("--input_dir",       required=True)
    p.add_argument("--output_dir",      required=True)
    p.add_argument("--device",          default="cuda",   choices=["cuda", "cpu"])
    p.add_argument("--batch_size",      type=int, default=DEFAULT_BATCH_SIZE,
                   help="VietOCR batch size (giảm xuống 8 nếu OOM)")
    p.add_argument("--workers",         type=int, default=DEFAULT_WORKERS)
    p.add_argument("--paddle_lang",     default="vi")
    p.add_argument("--vietocr_model",   default="vgg_transformer",
                   choices=["vgg_transformer", "vgg_seq2seq"])
    p.add_argument("--beamsearch",      action="store_true")
    p.add_argument("--no-resume",       dest="resume",   action="store_false")
    p.add_argument("--no-zip",          dest="auto_zip", action="store_false",
                   help="Bỏ qua bước zip sau OCR")
    p.add_argument("--gpu_clear_every", type=int, default=DEFAULT_GPU_CLEAR_EVERY)
    p.add_argument("--verbose",         action="store_true")
    args = p.parse_args()

    if args.device == "cuda":
        try:
            import torch
            if not torch.cuda.is_available():
                print("⚠️  CUDA not available → switching to CPU")
                args.device     = "cpu"
                args.batch_size = min(args.batch_size, 8)
            else:
                props = torch.cuda.get_device_properties(0)
                print(f"🎮 GPU: {props.name}  ({props.total_memory / 1e9:.1f} GB)")
        except ImportError:
            args.device = "cpu"

    run_folder(
        input_dir          = Path(args.input_dir),
        output_dir         = Path(args.output_dir),
        device             = args.device,
        batch_size         = args.batch_size,
        workers            = args.workers,
        paddle_lang        = args.paddle_lang,
        vietocr_model      = args.vietocr_model,
        vietocr_beamsearch = args.beamsearch,
        resume             = args.resume,
        auto_zip           = args.auto_zip,
        gpu_clear_every    = args.gpu_clear_every,
        verbose            = args.verbose,
    )


if __name__ == "__main__":
    main()
```
